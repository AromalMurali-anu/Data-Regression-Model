```python
x=df1.drop(columns=['Cooling Load'])     #deleting Cooling Load
```

```python
x.columns
```

```python
y=df1['Cooling Load']
```

```python
y
```

```python
from sklearn.feature_selection import SelectKBest ,f_regression  #importing SKLearn
```

```python
sk=SelectKBest(score_func=f_regression,k='all')
```

```python
x_new=sk.fit_transform(x,y)
```

```python
selected_columns=x.columns[sk.get_support()]
```

```python
selected_columns
```

```python
xselected=x[selected_columns]
```

```python
featurescore=pd.DataFrame({'features':x.columns,'score':sk.scores_
                          }).sort_values(by='score',ascending=False)
```

```python
featurescore
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
xscaled=ss.fit_transform(xselected)
```

```python
xscaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(xscaled,y,test_size=0.2,random_state=42)
```

```python
print("x_train shape:",x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
yperd_lr=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae=mean_absolute_error(y_test,yperd_lr)
mse=mean_squared_error(y_test,yperd_lr)
r2=r2_score(y_test,yperd_lr)
```

```python
print(mae)
print(mse)
print(r2)
```

```python
from sklearn.svm import SVR
```

```python
svr=SVR()
```

```python
svr.fit(x_train,y_train)
```

```python
yperd_svr=svr.predict(x_test)
```

```python
mae=mean_absolute_error(y_test,yperd_svr)
mse=mean_squared_error(y_test,yperd_svr)
r2=r2_score(y_test,yperd_svr)
```

```python
print(mae)
print(mse)
print(r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dr=DecisionTreeRegressor()
```

```python
dr.fit(x_train,y_train)
```

```python
yperd_dr=dr.predict(x_test)
```

```python
mae=mean_absolute_error(y_test,yperd_dr)
mse=mean_squared_error(y_test,yperd_dr)
r2=r2_score(y_test,yperd_dr)
```

```python
print(mae)
print(mse)
print(r2)
```
